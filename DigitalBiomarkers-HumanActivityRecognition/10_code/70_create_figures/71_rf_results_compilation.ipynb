```python
acc = pd.read_csv('../../../30_results/30_data_for_tableau/loocv_acc.csv')
phys = pd.read_csv('../../../30_results/30_data_for_tableau/loocv_phys.csv')
both = pd.read_csv('../../../30_results/30_data_for_tableau/loocv_both.csv')
```

```python
acc.columns = ['subject_id', 'ACC_accuracy', 'ACC_f1']
acc.head()
```

```python
phys.columns = ['subject_id', 'PHYS_accuracy', 'PHYS_f1']
phys.head()
```

```python
both.columns = ['subject_id', 'BOTH_accuracy','BOTH_f1']
both.head()
```

```python
df = pd.merge(acc, phys)
df = pd.merge(df, both)
```

```python
df = df[['subject_id','PHYS_accuracy','ACC_accuracy','BOTH_accuracy','PHYS_f1','ACC_f1','BOTH_f1']]
```

```python
df.head()
```

```python
df['BOTH-ACC_accuracy'] = df.apply(lambda row: row.BOTH_accuracy - row.ACC_accuracy,axis=1)
```

```python
df['BOTH-ACC_f1']=df.apply(lambda row: row.BOTH_f1 - row.ACC_f1,axis=1)
```

```python
df.head()
```

```python
df['BOTH-ACC_mean']= (df['BOTH-ACC_accuracy'] + df['BOTH-ACC_f1'])/2
```

```python
df = df.set_index('subject_id')
```

```python
df.sort_values(by='BOTH-ACC_mean',ascending = False)
```

```python
df.to_csv('rf_individual_metrics.csv')
```
