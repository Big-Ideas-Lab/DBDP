Imbalance of timepoints may or may not be likely to cause problems based on how we format inputs to our model. If the inputs are each timepoint or time windows of equal length (10 seconds for instance), then there will be an imbalance in the number of inputs between classes the model gets in training. This may cause the model to be biased in classifying activities that are more prevalent. It also brings up problems in model evaluation since scores like accuracy will be dominated by how well the prevalent classes are doing. 

Model bias may be attenuated by implementing specific parameter weighting schemes such as "class weighting". This is given as an option on Dataiku. Evaluation issues could be solved by looking past accuracy for more specific and robust measures.

Time-relevant data can be expressed in both windows and individual timepoints. Individual timepoints can use rolling methods to express local time data and windows are periods of time in and of themselves. Individual timepoints would have their values replaced with time-relevant ones. Summary statistics are needed to describe windows to be used in traditional ML. Each sensor needs to have one value, and that value could be something like the mean reading in that window. In deep learning models like LSTM, the window as a whole is considered for each classification, and thus summary statistics are not needed. Features generated by deep learning may be able to capture aspects of timeseries data that summary statistics and handmade timeseries features cannot. 

Another option is to have each round itself be an input to the model. However, in deep learning models, the varying lengths of rounds may be an issue. For instance, the model could learn to classify activity simply based on how long the activity lasts, which is not generalizable to data outside of this dataset. This is not a problem if summary statistics are used to represent the rounds, however it may be the case that the round length makes it such that a single summary statistic oversimplifies the complexity that happens during the entire round. Additionally, some models may require equivalently sized vectors as input across training instances.

If windows are chosen to be inputs to the model, using less data from the baseline and activity classes may be a good option for reducing imbalance. For instance, we could break up the activities into 30-second chunks and select 2 chunks of each activity per participant. Then, there is no class imbalance while maintaining equal representation of participants. This general principal can be applied to timepoints as well by taking a continuous minute long sample from the baseline and "activity" activities.

Using windows and timepoints seems more appealing for traditional ML methods since more localized time may be easier to express in summary statistics. Windows appear to be a great candidate for DL methods since window length can be normalized.

When using either windows or timepoints, one should try using a subset of the data they aleady have to reduce class imbalance. If this subset is sufficient for classification, creating time-series data through forcasting methods is not required. Not using forcasting methods is ideal since all of the data is actually collected. Data augmentation could be used to make new data by adding noise to already collected data and using the noise added window in addition to the original window. This may be problematic since there now would be an imbalance in redundancy and noise between activities. 

None of the problems of adding data are in subsetting data, so if there is enough data to build a good model after subsetting, subsetting should be used over adding data.

## Sensor Reading Distributions

This set of plots displays how sensor readings are distributed differently for separate activities. It appears that accelerometry and HR may be best in distinguishing our 4 activities from each other. 


```python
ac_list = ['Baseline', 'Type', 'DB', 'Activity' ]
sense_list = ['ACC1', 'ACC2', 'ACC3', 'TEMP', 'EDA', 'BVP', 'HR', 'Magnitude']

nrows = 8
ncols = 4
fig, ax = plt.subplots(nrows, ncols, figsize = (30,18))
colors = ['Blue','Orange','Green','Red']
for j in range(ncols):
    dfw = df1[df1["Activity"] == ac_list[j-1]] 
    color = colors[j]
    for i in range(nrows):
        if i == 0:
            ax[i,j].hist(dfw["ACC1"],40, range = [-150, 150], color = color)
        elif i == 1:
            ax[i,j].hist(dfw['ACC2'], 40, range = [-150, 150], color = color)
        elif i == 2:
            ax[i,j].hist(dfw['ACC3'], 40, range = [-150, 150], color = color)
        elif i == 3:
            ax[i,j].hist(dfw['TEMP'], 40, range = [20, 40], color = color)
        elif i == 4:
            ax[i,j].hist(dfw['EDA'], 50, range = [0, 80], color = color)
        elif i == 5:
            ax[i,j].hist(dfw['BVP'], 30, range = [-750, 750], color = color)
        elif i == 6:
            ax[i,j].hist(dfw['HR'], 50, range = [0, 180], color = color)
        else:
            ax[i,j].hist(dfw['Magnitude'], 30, range = [0, 150], color = color)
            
for i in range(ncols):
    ax[0,i].set_title(ac_list[i])
    
for j in range(nrows):
    ax[j,0].set(ylabel = sense_list[j])

plt.suptitle("Activity Type and Sensor Readings", size = 40, y = 0.93)
plt.show()
```

## Outliers by Activity
<a id="outact"></a>

Clearly, one would want to exclude sensor readings which have some malfunction as to not actually measure data correctly from the subject. Based on the valuues themselves, there is no way of knowing if the reading is due to a malfunction if the reading is in a similar range to what is normal physiologically. However, once values enter the physiologically impossible range, it is safe to say that those values are errors to be removed and either replaced with interpolated values, or excluded in total along with the other data points in its segment. 

Physiological impossibility also extends to how data is changing over time. For instance, heart rate changing by 40 bpm every second, but staying in a normal range, would be a malfunction even though each value taken by itself is a reasonable heart rate.

Assessing physiological impossibility requires domain knowledge of the sensor readings one is interpreting. For readings like heart rate and skin temperature, interpretability is relatively easy. A heart rate of 300 bmp and a skin temperature of 9 degrees celcius clearly indicates that either the subject is dead or there was a sensor malfunction. No subjects died during experimentation, so the assesment is straightforward. What is less straightforward is selecting boundaries of physiological possibility for these readings. Acceptable limits of physiology can be fuzzy.

Discussion with an expert is required to set boundaries for EDA, BVP, and ACC data. This becomes immediately obvious when one tries to think of what normal values for each of these readings are. What does an EDA of 10 mean? What does a heart rate of 10 mean? That is the difference in interpretability. It is innapropriate to exclude values just because they are relatively extreme. They must be supraphysiological. One cannot tell if a value is supraphysiological if that value doesn't have a physiological meaning to them.

Once these limits are set, there comes the question of what to do with these values. Do we interpolate them to acceptable values? How would we do that? Would we throw out that timepoint? Is throwing out points good practice when using windows?

Before answering any of these questions, let's take a look to see if any of these values actually exist.

## Outliers By Time
<a id="outtime"></a>

Here we can visualize where more extreme values for HR and TEMP come up over the course of one round of an activity (baseline) for an individual participant.

```python
df_baseline1 = df[df['Activity'] == 'Baseline 1']
df_baseline1['timedelta'] = df_baseline1.Time.astype(int).div(10**9).groupby(df['Subject_ID']).transform(lambda x: x.diff().fillna(0).cumsum())

# Rounded times
df_baseline1['timedelta_round1'] = df_baseline1['timedelta'].div(1).round(0) * 1
df_baseline1['timedelta_round2'] = df_baseline1['timedelta'].div(2).round(0) * 2
df_baseline1['timedelta_round5'] = df_baseline1['timedelta'].div(5).round(0) * 5
df_baseline1['timedelta_round10'] = df_baseline1['timedelta'].div(10).round(0) * 10
```

```python
plt.figure(figsize = (20,5))
sns.boxplot(df_baseline1['timedelta_round5'].astype(int), df_baseline1['TEMP'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (20,5))
sns.boxplot(df_baseline1['timedelta_round5'].astype(int), df_baseline1['HR'])
plt.xticks(rotation=90)
plt.show()
```

## Clustering on Sensor Summary Statistics
<a id="cluster"></a>

It's obvious that different people have different physiological makeups. Differences in physiological makeup may be important for classification since readings could look completely different for the same activity. A good classifier will be able to determine activity from multiple physiological profiles. This is an example of domain generalization. Each participant is a different source of data, perhaps a different domain. This clustering is done to see if any obvious general domains exist in our dataset. 

Summary statistics can be used as values to represent different domains to differentiate kinds of physiological profiles. For instance, mean heart rate is an interpretable value that makes sense for people to vary on. The mean heart rate of someone could potentially differentiate them being an athlete or a person who never exercises. The following clustering is done to see if any obvious groups of subjects exist as a particular "kind" of subject. 

If the clustering does not work, that does not mean each subject is not a different domain. It just means that subjects do not fall into easily differentiable groups. If they do fall into differentiable groups, we could look for class imabalances there. Additionally, differentiable groups may further motivate DL techniques that prioritize domain generalization.

```python
summary = pd.DataFrame()

for i in range(1,57):
    if i < 10:
        summary[i] = a1.loc[a1['Subject_ID'] == '19-00'+str(i)].TEMP.describe()     
    if i >= 10:
        summary[i] = a1.loc[a1['Subject_ID'] == '19-0'+str(i)].TEMP.describe()
summary = summary.transpose()
summary = summary.drop('count', axis =1)
summary.head()
```

```python
plt.scatter(summary['mean'],summary['std'])
```

```python
plt.scatter(summary['50%'],summary['std'])
```

```python
test = summary[['mean','std']]

```

```python
import sklearn
from sklearn.cluster import KMeans
k = 5
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(test)

plt.scatter(test['mean'], test['std'], c=y_pred, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python


h = .02
x_min = test['mean'].min() - 1
y_min = test['std'].min() - 1
x_max = test['mean'].max() + 1
y_max = test['std'].max() + 1

xx, yy = np.meshgrid(np.arange(x_min,x_max,h), np.arange(y_min, y_max, h))
  
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])







Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
       extent=(xx.min(), xx.max(), yy.min(), yy.max()),
       cmap=plt.cm.Paired,
       aspect='auto', origin='lower')

plt.plot(test['mean'], test['std'], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
        marker='x', s=169, linewidths=3,
        color='w', zorder=10)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

There do not appear to be obvious clusters in the data that would easily group participants. Fortunately then, we do not have to worry about class imbalances for different "kinds" of physiological profiles.

We do still have to worry about domain generalization though. Each participant is still their own "domain" because it is an undeniable fact that different physiological profiles exist. Therefore any testing set of a model must include participants (domains) the model did not train on. Further, specific techniques may be implemented in DL model creation to emphasize domain generalizability. 

## 3D Animation of Accelerometer

After attempting to plot all 3 accelerometry axes on 3d plots, it was evident that they were very uninterpretable. Animating these plots allows one to see the change over time of the accelerometry measures, effectively representing the movement in three dimensions over that interval.

```python
from mpl_toolkits import mplot3d
```

```python
#I would recommend making interval 50 or below. This will shorten the video as much as possible.
#Subject and activity must be strings.
#This can run upwards of 5 minutes.

def animation3d(df, subject, activity, interval):
    fig = plt.figure()
    ax = p3.Axes3D(fig)
    #plt.clf()
    
    ax.clear()
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    
    plt.rcParams['animation.html'] = 'html5'
    plt.rcParams['figure.figsize'] = [30, 18]

    df_new = df[(df['Subject_ID'] == subject) & (df['Activity'] == activity)]
    
    x1 = df_new['ACC1']
    y1 = df_new['ACC2']
    z1 = df_new['ACC3']
    
    lns = []
    for i in range(len(df_new['ACC1'])):
        ln1 = ax.plot(x1[:i], y1[:i], z1[:i], 'o-', color='steelblue')
        lns.append(ln1)

    line_ani = animation.ArtistAnimation(fig, lns, interval=interval, blit=False)
    return line_ani
```
